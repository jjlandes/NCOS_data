```python
import pandas as pd
```

```python
data = pd.read_csv('Summer research compiled data sheet - Sheet1.csv')
```

```python
salinity = data.loc[:, "EC (mS)"]
```

```python
import matplotlib.pyplot as plt
```

```python
dry_veg = data.loc[:, "Vegetation dry weight"]
```

```python
plt.scatter(salinity, dry_veg)
plt.xlabel("Salinity (mS)")
plt.ylabel("Vegetation dry weight (g)")

# Peason's r
salinity.corr(dry_veg)
dry_veg.corr(salinity)

#Spearman's rho
salinity.corr(dry_veg, method='spearman')

# Kendall's tau
salinity.corr(dry_veg, method='kendall')
```

```python
elevation = data.loc[:, "Elevation"]
sed_accum = data. loc[:, "Average sediment accumulation"]
```

```python
plt.scatter(elevation, sed_accum)
# take the average of each of the four elevations and see if there is a strong correlation then
```

```python
clay = data.loc[:, "% Clay"]
```

```python
plt.scatter(clay, sed_accum)

clay.corr(sed_accum)
sed_accum.corr(clay)
```

```python
plt.scatter(clay, salinity)
plt.xlabel('Percent clay')
plt.ylabel('Salinity (mS)')

clay.corr(salinity)
salinity.corr(clay)
```
